Ok, let us get started.

First things first - let us get some data to play around with. We are going to use a data set about the pricing of diamonds.

```python
data = sns.load_dataset("diamonds")
data["cut"] = data["cut"].cat.codes
data["color"] = data["color"].cat.codes
y = data["price"]
X = data.drop(["price", "clarity"], axis=1)

X.head()
```

```python
def div(arg: str):
  print()
  print(f"============== {arg} ==============")
```

Ok, so we have a numeric table. Let's have a quick look at the data and familiarize ourselves a little with Pandas.
- What are the dtypes?                                         *(hint: use .info())*
- How many data points have a carat value of 0.3 or higher?    
- What is the average price for each cut?                      *(hint: use .groupby())*




```python
## get the above statistics
div("dtypes")
print(data.info())
div("Number of data points with carat >= 3")
print(f"{(X["carat"] >= 0.3).sum()}")
div("Price for each cut:")
print(data.groupby("cut")["price"].mean())
```

Before we fit a model we always *normalize* the Input data X and *standardize* the output data y. Go ahead and do that now.

```python
## normalize / standardize the output data:
x_scaler = sklearn.preprocessing.StandardScaler()
y_scaler = sklearn.preprocessing.StandardScaler()

X_scaled = x_scaler.fit_transform(X)
y_scaled = y_scaler.fit_transform(y.to_frame()).ravel()
```

The next thing we do is splitting the data into a train and a test set. Use sklearn.model_selection.train_test_split().
- How large is the data set?
- What would be a reasonable train-test split?

```python
from sklearn.model_selection import train_test_split
print(f"Dataset size: {len(X)}")

(
    X_train,
    X_test,
    y_train,
    y_test
) = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

print("Reasonable train-test split is usually 10-20%, depending on datasize")
```

Alright, so we have some data - let us make a linear regression model! You can code your own or you can use the LinearRegressor from sklearn.linear_model - whichever you prefer. Regardless, the goal is simply to fit a linear regressor to the training data and then predict the output for the test data.

```python
## Create and train your linear regressor
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

## Predict the output of the test data
y_pred = model.predict(X_test)
```

Ok, once we have made our prediction, let us make a simple plot to see what is going on. Our data is multidimensional, so our predictor will be a multidimensional hyperplane which we cannot plot. Instead, let us plot the predicted prices agains the true prices to see how we did. Here, a scatter plot would be in order.

```python
## Make the parity plot described above
plt.scatter(y_test, y_pred)
plt.xlabel("True prices")
plt.ylabel("Predicted")
plt.show()
```

Alright, we are nearing the end of of this small task.
- What loss function did you use for the linear regressor?

As the last thing, calculate the training error and the test error for your regressor.

**Answer:** Linear regression uses L2 loss function

```python
## What is your training and test error?
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_error = sklearn.metrics.mean_squared_error(y_train, y_train_pred)
test_error = sklearn.metrics.mean_squared_error(y_test, y_test_pred)

print(f"Training error: {train_error}")
print(f"Test error: {test_error}")
```

Ok, so what could you do to make this model better?
- Do we have a bias problem or a variance problem?

To help find out, plot the training loss and the test loss against the training set size (train and evaluate the model on increasing amount of data).
- What can we say about the results?
- Name a few alteratives to resolve it.

```python
#just retraining all models is easier
train_sizes = np.linspace(0.1, 1.0, 10)
train_losses = []
test_losses = []

for frac in train_sizes:
    m = int(frac * len(X_train))

    X_sub = X_train[:m]
    y_sub = y_train[:m]

    model = sklearn.linear_model.LinearRegression()
    model.fit(X_sub, y_sub)

    train_loss = np.mean((y_sub - model.predict(X_sub))**2)
    test_loss = np.mean((y_test - model.predict(X_test))**2)

    train_losses.append(train_loss)
    test_losses.append(test_loss)
```

```python
## Plot the train and test error against increasing amount of training data?
plt.plot(train_sizes * len(X_train), train_losses, label="Training loss")
plt.plot(train_sizes * len(X_train), test_losses, label="Test loss")
plt.xlabel("Training set size")
plt.ylabel("MSE")
plt.legend()
plt.show()
```

**ANSWER**: Probably bias problem, train/test error is very close so model doesn't overfit

After graph visualization:

**ANSWER:** Model is probably to simple to capture connections in the data. Price/carat is probably linear, so a nonlinear model would be better in this scenario.

The model generalizes well but is not complex enough (seen by train/test being very close together).
